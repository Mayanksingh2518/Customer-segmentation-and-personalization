```python
import pandas as pd
import datetime as dt

# Load the dataset with appropriate encoding
data = pd.read_csv("OnlineRetail.csv", encoding='latin1')

# Convert 'InvoiceDate' to datetime
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])

# Create 'TotalPrice' column
data['TotalPrice'] = data['Quantity'] * data['UnitPrice']

# Create snapshot date for recency calculation
snapshot_date = data['InvoiceDate'].max() + dt.timedelta(days=1)

# Create RFM table
rfm = data.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo': 'count',
    'TotalPrice': 'sum'
})

# Rename columns for RFM
rfm.rename(columns={'InvoiceDate': 'Recency', 'InvoiceNo': 'Frequency', 'TotalPrice': 'Monetary'}, inplace=True)

# Display the first few rows of the RFM table
rfm.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Normalize RFM values
scaler = StandardScaler()
rfm_normalized = scaler.fit_transform(rfm)

# Apply K-means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
rfm['Cluster'] = kmeans.fit_predict(rfm_normalized)

# Visualize clusters
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x='Recency', y='Frequency', hue='Cluster', data=rfm, palette='viridis')
plt.title('Customer Segments based on RFM')
plt.show()
```

```python
# Create a detailed report with the correct aggregation
report = rfm.groupby('Cluster').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': 'mean',
    'Recency': 'count'  # This counts the number of customers in each cluster
}).rename(columns={'Recency': 'CustomerCount'}).reset_index()

print(report)

# Visualize monetary value by cluster
sns.boxplot(x='Cluster', y='Monetary', data=rfm)
plt.title('Monetary Value by Customer Segment')
plt.show()
```

```python
import pandas as pd
import datetime as dt
from sklearn.cluster import KMeans

# Load the dataset with appropriate encoding
data = pd.read_csv("OnlineRetail.csv", encoding='latin1')

# Convert 'InvoiceDate' to datetime
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])

# Create 'TotalPrice' column
data['TotalPrice'] = data['Quantity'] * data['UnitPrice']

# Create snapshot date for recency calculation
snapshot_date = data['InvoiceDate'].max() + dt.timedelta(days=1)

# Create RFM table
rfm = data.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo': 'count',
    'TotalPrice': 'sum'
})

# Rename columns for RFM
rfm.rename(columns={'InvoiceDate': 'Recency', 'InvoiceNo': 'Frequency', 'TotalPrice': 'Monetary'}, inplace=True)

# Assign clusters (example using KMeans)
kmeans = KMeans(n_clusters=4)
rfm['Cluster'] = kmeans.fit_predict(rfm[['Recency', 'Frequency', 'Monetary']])

# Create detailed report
report = rfm.groupby('Cluster').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': 'mean',
    'Cluster': 'count'
}).rename(columns={'Cluster': 'CustomerCount'}).reset_index()

print(report)
```
